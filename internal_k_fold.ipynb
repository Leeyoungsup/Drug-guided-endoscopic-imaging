```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold
```

```python
df=pd.read_csv('../../data/mteg_data/lateral_internal/label.csv')
kf = KFold(n_splits = 5, shuffle = True, random_state = 50)
x = np.array(df)
y = df['OTE 원인']
```

```python
count=0
for train_index, test_index in kf.split(x):
    count+=1
    X_train, X_test = x[train_index], x[test_index]
    y_train, y_test = y[train_index], y[test_index]
    train_df=pd.DataFrame(X_train, columns=df.columns)
    test_df=pd.DataFrame(X_test, columns=df.columns)
    train_df.to_csv('../../data/mteg_data/lateral_internal/'+str(count)+'-fold_train.csv', index=False)
    test_df.to_csv('../../data/mteg_data/lateral_internal/'+str(count)+'-fold_test.csv', index=False)
    
    
```
